# Cleaning data and Explore dataset
This dataset export from Weple money application that I recorded my spending since 2018.

## Import libraries

```python
# Importing the required packages
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

## Read dataset

```python
df = pd.read_csv("/content/drive/MyDrive/weple_2565-01-16.csv", encoding='utf-16')
df.head() # preview first 5 rows of dataset
```

The error message I received when reading this CSV file was `"UnicodeDecodeError: 'utf-8' codec can't decode byte 0xff in position 0: incorrect start byte."` It indicates that when Pandas attempts to decode a series of bytes into a string, it runs into a series of bytes that is not permitted by the `"utf-8"` encoding. I tried the `"utf-16"` encoding, and it worked. When we set `encoding='utf-16'` pandas will read the CSV file using this encoding and convert the data to a DataFrame object. With alternative encodings, non-ASCII characters in the data might not be properly interpreted. This enables pandas to handle non-ASCII characters in the data accurately.

## Manipulate dataset
Clean the data to determine which columns are redundant or unnecessary

```python
# Remove white space from columns name
df.columns = df.columns.str.replace(' ', '')
```

```python
# The white space has disappeared
df.columns
```

```python
# Cleanse the data to determine which columns are redundant or unnecessary
# Drop unwanted columns
col_list = ['사용자', '하위분류', 'Card', 'Memo', 'Payment', 'Text']
df = df.drop(col_list, axis=1)

df.head()
```

```python
# We found the duplicate rows and then we will drop the duplicates
print('Before:', df.shape)
df = df.drop_duplicates(keep='first')
print('After:', df.shape)
```

```python
# Dropping the rows that contain a specific string
# We will drop the income rows because we want to analyze only expense
print('Before:', df.shape)
df = df[df["수입/지출"].str.contains("수입") == False]
print('After:', df.shape)
```

```python
# Changing the Date column to DateTime column
df.Date = pd.to_datetime(df['Date'])
```

Because the year in this dataset is out of bounds, I received an error message: "OutOfBoundsDatetime: Out of bounds nanosecond timestamp: 2565-01-15 00:00:00" when I changed date columns to DateTime columns. To fix this error, I will check that timestamps are allowed and create a new column to manage the datetime.

```python
# Check the date range in dataset
print('Min date :' + str(df['Date'].min()))
print('Max date :' + str(df['Date'].max()))
```

```python
# Display minimum timestamp allowed
print("Minimum time stamp allowed is :", pd.Timestamp.min)
# Display maximum timestamp allowed
print("Maximum time stamp allowed is :",pd.Timestamp.max)
```

```python
# Creat new columns for manage date column
df[["year", "month", "day"]] = df["Date"].str.split("-", expand = True)
df
```

```python
df['year'] = df['year'].astype(int)
df['year'] = df['year'] - 543
print(df['year'])
```

```python
df.head()
```

```python
df['new_Date'] = pd.to_datetime(df[['year', 'month', 'day']])
df.head()
```

```python
# Convert year,month,day columns datatype from charactor to numeric.
df.year = pd.to_numeric(df.year)
df.month = pd.to_numeric(df.month)
df.day = pd.to_numeric(df.day)
```

```python
df.info()
```

**See top 10 category spending.**

```python
top10 = df[["Category", "Amount"]].groupby("Category", as_index = False).sum().sort_values(by = "Amount", ascending = False).reset_index().head(10)
top10
```

```python
sns.set_theme(style="whitegrid", palette='pastel')
plt.figure(figsize = (7, 5))
sns.barplot(x = 'Amount', y = 'Category', data = top10)
plt.title("Top 10 spending by category")
```

## Amount
Daily Amount.

```python
daily_amount = pd.DataFrame(df.groupby("Date_new").Amount.sum())
daily_amount
```

```python
fig, ax = plt.subplots(figsize = (15, 5))
ax.plot(daily_amount.index, daily_amount.Amount)
plt.title("Transaction volume")
plt.xlabel("Date")
plt.ylabel("Transaction volume")
```

```python
#month_amount = df.groupby(['month', 'Category'])['Amount'].sum().reset_index()
month_amount = pd.DataFrame(df.groupby("month").Amount.sum())
month_amount
```
